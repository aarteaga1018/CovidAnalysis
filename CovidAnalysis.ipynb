```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

```python
#get data
corona_dataset_csv = pd.read_csv("covid19_deaths_dataset.csv")
```

```python
#view dataframe
corona_dataset_csv.head()
```

```python
#get shape of dataframe
corona_dataset_csv.shape
```

```python
#delete useless rows 
corona_dataset_csv.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
corona_dataset_csv.head()
```

```python
#get name of column we want to perform groupby on and sum rows 
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.tail(15)
```

```python
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc["US"].plot()
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
#reviewing data curves in U
corona_dataset_aggregated.loc["US"][40:45].plot()
```

```python
corona_dataset_aggregated.loc["US"].diff().plot()
plt.legend()
```

```python
#max reporting of day of cases reported 
corona_dataset_aggregated.loc["US"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
#narrowing down data
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

```python
#importing happiness data
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report_csv.head()
```

```python
useless_cols = ["Overall rank","Generosity","Perceptions of corruption","Score"]
```

```python
#drop useless columns
happiness_report_csv.drop(useless_cols,axis=1,inplace=True)
happiness_report_csv.head()
```

```python
#make Country/region the index
happiness_report_csv.set_index("Country or region",inplace=True)
happiness_report_csv.head()
```

```python
#join the data
data = corona_data.join(happiness_report_csv,how="inner")
data.head()
```

```python
#finding an correlations
data.corr()
```

```python
#finding additonal info
data.describe()
```

```python
x = data["GDP per capita"]
y = data["max_infection_rate"]
sns.scatterplot(x, np.log(y))
```
